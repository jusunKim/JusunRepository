```python
import re
```

```python
# 불용어(stop word) : 분석에 필요없는 단어. 별의미없는거
# 그런 건 분석에서 제외시킨다
```

```python
# 한글이 아닌 모든 문자를 공백으로 변환하자
moon = re.sub('[^가-힣]',' ',moon)
moon
```

```python
# 분석에 사용하지 않을 단어 불용어를 배열에 담아보자
stop_word = ['나라','국민','우리','대통령','대한민국','우리나라']
for word in stop_word:
  moon = re.sub(word," ",moon)
moon
```

```python
import konlpy
```

```python
hannanum =konlpy.tag.Hannanum()
```

```python
# 한나눔한테 명사만 뽑아달라고 시키기
hannanum.nouns("대한민국의 영토는 한반도와 그 부속도서로 한다")
```

```python
# 대통령 연설문으로부터 명사만 추출해보자
nouns = hannanum.nouns(moon)
nouns #얘 자료형이 list라서 데이터프레임으로 바꿔줘야대
```

```python
# 추출된 명사를 데이터프레임으로 만들자
df_word = pd.DataFrame({'word':nouns})
df_word
```

```python
#단어의 길이가 2글자 이상인 것만 뽑기
df_word['count'] = df_word['word'].str.len()
df_word = df_word[df_word['count']>2]
df_word
```
