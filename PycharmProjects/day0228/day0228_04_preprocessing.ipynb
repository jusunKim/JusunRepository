```python
student = exam.copy()
#student에 수학, 과학, 영어 합한 total파생변수 추가
```

```python
student['total'] = student['math']+student['english']+student['science']
student
```

```python
# 파생변수 추가를 위한 판다스의 assign() 이용하기
# df.assign(변수명=값)
```

```python
student = exam.copy()
student = student.assign(total=student['math']+student['english']+student['science'])
student
```

```python
# 한번에 여러개의 파생변수 만들기
# assign(변수1 = 값1, 변수2=값2...)
student = exam.copy()
student = student.assign(total=student['math']+student['english']+student['science'],mean =(student['math']+student['english']+student['science'])/3 )
student
```

```python
student = exam.copy()
student = student.assign(total=lambda x: x['math']+x['english']+x['science'],mean =lambda y: y['total']/3 )
student
```
