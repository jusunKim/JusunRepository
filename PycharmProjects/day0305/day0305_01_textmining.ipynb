```python
import re
```

```python
moon = re.sub('[^가-힣]',' ',moon)
stop_word = ['나라','국민','우리','대통령','우리나라','대한민국']
for word in stop_word:
  moon = re.sub(word," ",moon)
moon
```

```python
# 한국어 형태소 분석을 위한 konlpy 설치
!curl -s https://raw.githubusercontent.com/teddylee777/machine-learning/master/99-Misc/01-Colab/mecab-colab.sh | bash
```

```python
import konlpy
```

```python
# 형태소 분석을 위한 한나눔 객체 생성
hannanum = konlpy.tag.Hannanum()
# 한나눔 객체를 통해 연설문에서 명사만 추출하기
nouns = hannanum.nouns(moon)
nouns
```

```python
# 추출된 명사를 데이터프레임으로 만들기
df_word = pd.DataFrame({
    'word':nouns
})
df_word
```

```python
# 각 단어 길이 구해서 파생변수 count추가
df_word['count'] = df_word['word'].str.len()
df_word
```

```python
df_word = df_word.query('count>=2')
df_word
```

```python
# 각 단어별 빈도표 만들기
df_word = df_word.groupby('word',as_index=False).agg(n=('word','count')).sort_values('n',ascending=False)
df_word
```

```python
#상위20개 추출해 막대그래프
top20 = df_word.head(20)
```

```python
# 워드클라우드 만들기: 단어별 빈도표 만들고(데이터프레임) 걔를 딕셔너리로 만들어야
# word를 key로 만들어야 하고 n(빈도)를 value로 만들어야댐
dic_word = df_word.set_index('word').to_dict()['n']
```

```python
dic_word
```

```python
# 워드클라우드 패키지 읽어들이기
from wordcloud import WordCloud
```

```python
# 글씨 경로를 font라는 변수에넣기
font = "/content/gdrive/My Drive/DoHyeon-Regular.ttf"

#워드클라우드 객체 생성
wc = WordCloud(font_path=font, width=400, height=400, background_color='white')

#위 워드클라우드 객체에 데이터를 설정하기
img_wordcloud = wc.generate_from_frequencies(dic_word)

#생성된 워드클라우드 이미지를 matplotlib을 통해 화면에 그려줌
import matplotlib.pyplot as plt
plt.figure(figsize=(10,10)) #도화지 크기 설정
plt.axis('off') #축은 표시하지마
plt.imshow(img_wordcloud) #이미지 출력
```

```python
#워드클라우드를 원하는 모양으로 만들기
# 이미지 준비해서 그 모양으로 워드클라우드 출력할 수 잇삼

#이미지 처리 위한 패키지 포함시키기
import PIL

#파이썬 이미지 처리 모듈인 PIL통해서 이미지 읽어들인다
icon = PIL.Image.open('/content/gdrive/My Drive/cloud.png')

#마스크이미지 생성을 위해 numpy 포함시켜야댄다(난 아까 위에서 이미 함)

#PIL을 통해 icon파일의 크기와 동일한 배경이 흰색인 이미지 생성
img = PIL.Image.new('RGB',icon.size,(255,255,255))

#마스크이미지 생성을 위해 icon을 icon에 복사한다???>?
img.paste(icon,icon)

#img를 배열로 만들어줌
img = np.array(img)
img

```

```python
# 생성된 마스크이미지 img를 갖고 워드클라우드 생성하기
wc = WordCloud(
    font_path= font, width=400, height=400,background_color="white",mask = img
)
```

```python
#워드클라우드 객체에 데이터 설정하기
img_wordcloud = wc.generate_from_frequencies(dic_word)

#이미지 그리기
plt.figure(figsize=(10,10))
plt.axis('off')
plt.imshow(img_wordcloud)
```
