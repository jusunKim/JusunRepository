```python
mpg = pd.read_csv("gdrive/My Drive/mpg.csv")
```

```python
df = mpg.copy()
```

```python
#1) 배기량에 따라 고속도로 연비가 다른지(displ배기량이 4이하인자동차와 5이상인 차 중 어떤 게 hwy가 더 높을까)
df1 = df[df['displ']<=4]
df2 = df[df['displ']>=5]

print(df1['hwy'].mean()) #25.96319018404908
# print(df.query("displ<=4")['hwy'].mean()) #25.96319018404908
print(df2['hwy'].mean()) #18.07894736842105
# print(df.query("displ>=5")['hwy'].mean()) #18.07894736842105
```

```python
#2.audi와 toyota 중 어느 manufacturer의 cty평균이 높을까
# df2 = df.query('manufacturer in ["audi","toyota"]').groupby('manufacturer',as_index=False).agg(cty_mean=('cty','mean'))
df2 = df.query('manufacturer in ["audi","toyota"]').pivot_table(values='cty',index='manufacturer') #피벗테이블로보기
df2
```

```python
# 3.셰보레, 포드, 혼다의 고속도로 연비 평균
df3 = df.query('manufacturer in ["chevrolet", "ford", "honda"]')['hwy'].mean()
#방법2. 따로 가져와서 붙인 다음에 평균 구하기
a = df[df['manufacturer']=='chevrolet']['hwy']
b = df[df['manufacturer']=='ford']['hwy']
c = df[df['manufacturer']=='honda']['hwy']
df3 = pd.concat([a,b,c]).mean()
df3
```

```python
# mpg에서 카테고리, 도시연비 변수만 추출해서 새 데이터를 만들어 보
df4 = df[['category','cty']]
df4
```

```python
#5위에서 뽑은 데이터에서 카테고리가 suv인 거랑 compact인 것 중 뭐가 cty가 평균이 높은지 알아보
# df5 = df4.query('category in ["suv","compact"]').groupby('category',as_index=False).agg(mean_cty=('cty','mean'))
#피벗테이블로 하기
df5 = df4.query('category in ["suv","compact"]').pivot_table('cty',index='category')
df5
```

```python
# 6. mpg 복사본을 만들고 cty와 hwy를 더한 합산 연비 변수를 추가합니다.
# df['total'] = (df['cty']+df['hwy'])
#assign함수 이용하기
df = df.assign(total=df['cty']+mpg['hwy'])
df
```

```python
# df['avg'] = df['total']/2
df = df.assign(avg = df['total']/2)
df
```

```python
# 8. 평균연비변수가 가장 높은 자동차 3종을 출력합니다.
df.sort_values('avg',ascending=False).head(3)
```

```python
# 9. 6~8번을 하나로 연결된 pandas 구문으로 만들어 봅니다.
df.copy().assign(total=df['cty']+mpg['hwy'],avg =( df['cty']+mpg['hwy'])/2)\
          .sort_values('avg',ascending=False).head(3)
```

```python
# 10. mpg데이터의 catetory는 자동차를 특징에 따라 suv, compact등 일곱 종류로 분류합니다.
# 어떤 차종의 도시연비가 높은지 비교하기 위하여 catetory별 cty의 평균을 구해 봅니다.
df.groupby('category',as_index=False).agg(mean=('cty','mean'))
```

```python
#피ㅓㅄ테이블로 해보자
df.pivot_table('cty',index='category').sort_values('cty') #피벗테이블도 sort_values가 된다
```

```python
# 13. 어떤 회사에서 compact 차종을 가장 많이 생산하는지 알아 봅니다.
df.query('category=="compact"').groupby('manufacturer').agg(n=('category','count')).sort_values('n',ascending=False)
```
