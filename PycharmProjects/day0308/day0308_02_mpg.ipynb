```python
# 연습1) 연료종류별 가격 정보를 df로 만들어보자
flprice = pd.DataFrame({
    'fl':['c','d','e','p','r'],
    'price_fl': [2.35, 2.38, 2.11, 2.76, 2.11]
})
flprice
```

```python
# 연습2)  mpg데이터에는 연료 정류를 나타내는 fl은 있지만 가격은 없음. 1에서 만든 데이터 이용해서
# mpg에 price_fl 변수를 추가
df = pd.merge(df,flprice)
# df = df.merge(flprice) 이렇게해도
df
```
